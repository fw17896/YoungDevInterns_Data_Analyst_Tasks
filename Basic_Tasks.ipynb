Load a dataset into Excel or Google Sheets.

```python
import pandas as pd
df = pd.read_csv("Global YouTube Statistics.csv" , delimiter = '\t' , encoding = "latin-1")

df = pd.read_csv("Global YouTube Statistics.csv" , nrows=100, encoding = "latin-1")
```

Identify and delete duplicate rows.

```python
# Remove duplicate rows
df_cleaned = df.drop_duplicates()
```

 Fill in or remove missing values.

```python
# Check missing values
print(f"Total duplicate rows: {df.duplicated().sum()}")

# Remove duplicates
df_cleaned = df.drop_duplicates()
print(f"After removing duplicates: {df_cleaned.shape}")
```

```python
# Fill missing numerical values with mean
for col in df_cleaned.select_dtypes(include=['number']).columns:
    df_cleaned[col].fillna(df_cleaned[col].mean(), inplace=True)
```

```python
# Fill missing categorical values with mode
df_cleaned.fillna(df_cleaned.mode().iloc[0], inplace=True)
```

**Task 2**

---


 Visualize data using bar plot.

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Top YouTubers by Subscribers
top_youtubers = df.nlargest(10, 'subscribers')[['Youtuber', 'subscribers']]
print("Top 10 YouTubers by Subscribers:\n", top_youtubers)

plt.figure(figsize=(10, 6))
plt.barh(top_youtubers['Youtuber'], top_youtubers['subscribers'], color='green')
plt.xlabel('Subscribers (in millions)')
plt.title('Top 10 YouTubers by Subscribers')
plt.show()
```

 Visualize data using pie charts.

```python
# Count of each category
category_counts = df_cleaned["category"].value_counts()

# Plot pie chart
plt.figure(figsize=(8,8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Distribution of YouTube Categories")
plt.show()
```

Find the mean, median, and standard deviation of a dataset.

```python
# Mean, Median, Std of Subscribers & Video Views
mean_subscribers = df_cleaned["subscribers"].mean()
median_subscribers = df_cleaned["subscribers"].median()
std_subscribers = df_cleaned["subscribers"].std()

mean_views = df_cleaned["video views"].mean()
median_views = df_cleaned["video views"].median()
std_views = df_cleaned["video views"].std()

print(f"Subscribers - Mean: {mean_subscribers}, Median: {median_subscribers}, Std Dev: {std_subscribers}")
print(f"Video Views - Mean: {mean_views}, Median: {median_views}, Std Dev: {std_views}")
```
